```python
# creating data frame of Meta Data key
alpha_frame = pd.DataFrame(columns = ['Information', 'Symbol', 'Last_Refreshed', 'Interval', 'Output_Size', 'Time_Zone'])
```

```python
def get_details(trade_info):
  try:
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&apikey=8WAK8FNJNSTN111M'
    response = requests.get(url)
    alpha_vantage = response.json
    return alpha_vantage['Information'], alpha_vantage['Symbol'], alpha_vantage['Last_Refreshed'], alpha_vantage['Interval'], alpha_vantage['Output_Size'], alpha_vantage['Time_Zone']
  except:
    return 'Error', np.NAN, np.NAN, np.NAN
```

```python
for trade_info in range(1, 50):
 alpha_frame.loc[trade_info] = get_details(trade_info)
```

```python
import pandas as pd

# upload dataset
project = pd.read_csv('project_dataset.csv')

```

```python
# removing writtenby from column

project['author'] = project['author'].str.replace('Writtenby:', '')

```

```python
# add space to first and surname
project['author'] = project['author'].str.split(r'(?=[A-Z])').str.join(' ')
```

```python
# clean narrator column as I did the author column

project['narrator'] = project['narrator'].str.replace('Narratedby:', '')

project['narrator'] = project['narrator'].str.split(r'(?=[A-Z])').str.join(' ')

```

```python
# cleaning the date format
project['releasedate'] = pd.to_datetime(project['releasedate'], format='%d-%m-%y')
```

```python
# create new field minutes and change time to minutes in the field
def change_time(time):
    if time == 'Less than 1 minute':
      return 0
    time = time.replace('hrs', 'hr').replace('mins', 'min')
    my_lists = time.split('and')

    hours = 0
    minutes = 0

    for my_list in my_lists:
        my_list = my_list.strip()
        if 'hr' in my_list:
            hours = int(my_list.split('hr')[0])
        if 'min' in my_list:
            minutes = int(my_list.split('min')[0])
    return hours * 60 + minutes


```

```python
# creating rating field

project['rating'] = project['stars'].str.split().str[0].replace('Not', 0).astype(float).round().astype(int)



```

```python
# applying the changes to minutes field
project['minutes'] = project['time'].apply (change_time)
```

```python
import re

def number_of_ratings_col(stars):
  rate = 0
  if str(stars).lower().strip() != 'not rated yet':
    rate = re.split('stars',str(stars))[1]
    rate = int(re.sub("[^0-9]", "",rate))
  else:
    rate = 0

  return rate

project['number_of_ratings'] = project['stars'].apply(number_of_ratings_col)

```

```python
# turn price into float column

def change_price(n):
  n = n.replace(',', '')
  if n == 'Free':
    return 0.0
  else:
    return float(n)

```

```python
project['price'] = project['price'].apply(change_price)
```

```python
project.to_csv('project_dataset_clean.csv', index=False)
```

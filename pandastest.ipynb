```python
import pandas as pd

# upload dataset
project = pd.read_csv('project_dataset.csv')

project.head()
```

```python
# this returns the bottom 5 data in the csv file

project.tail(5)
```

```python
# showing one column

project['author']
```

```python
# removing writtenby from column

project['author'] = project['author'].str.replace('Writtenby:', '')

```

```python
# check to see that it's gone

project['author']
```

```python
# add space to first and surname
project['author'] = project['author'].str.split(r'(?=[A-Z])').str.join(' ')
```

```python
# now test it out to see that it's worked
project['author']
```

```python
# show narrator column
project['narrator']
```

```python
# clean narrator column as I did the author column

project['narrator'] = project['narrator'].str.replace('Narratedby:', '')

project['narrator'] = project['narrator'].str.split(r'(?=[A-Z])').str.join(' ')

```

```python
# test output after previous steps
project['narrator']
```

```python
# cleaning the date format
project['releasedate'] = pd.to_datetime(project['releasedate'], format='')
```

```python
# check releasedate DATE has changed

project['releasedate']
```

```python
# create new field minutes and change time to minutes in the field
project['minutes'] = project['time'] * 60
```
